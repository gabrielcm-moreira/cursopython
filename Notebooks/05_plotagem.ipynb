Plote em mesmo gráfico um modelo gaussiano e um modelo exponencial de variograma com o mesmo alcance e mesmo patamar

```python
c_0 = 0
c = 10
r = 10
h = np.arange(0,100,1)
for i in range(10):
    y = c_0 + c*(1 - np.exp(-h / r))
for i in range(10):
    y2 = c_0 + c*(1 - np.exp(-h**2 / r**2))
plt.plot(h,y, label='var_exponencial')
plt.plot(h,y2, label='var_gaussiano')
plt.title('Variogramas')
plt.xlabel('Distância')
plt.ylabel('Gamma')
plt.legend()
```

# Exercício grids

Crie uma função que recebe um array de teores e um dicionário com os parâmetros do grid e plota.
 

```python
t = np.random.randn(20*10)
g = {'nx':20, 'ny':10, 'ox':0, 'oy':0, 'sx':5, 'sy':10}
```

```python
def pixelplot(teores,grid):     
    teores = teores.reshape(grid['ny'],grid['nx']) #Por convenção, a primeira entrada é o número de linhas, a segunda, de colunas
    plt.imshow(teores)
    plt.show()
```

```python
pixelplot(t,g)
```

# Exercício subplots

Plote um mapa de amostras ao lado de um grid.

```python
coordsx = np.linspace(0,10,10)
coordsy = np.linspace(0,10,10)
fig, axes = plt.subplots(1,2)
subfig = axes[0].scatter(coordsx, coordsy, c = np.random.randn(10))
axes[1].imshow(np.random.randn(10,10))
fig.colorbar(subfig)
plt.show()
```
