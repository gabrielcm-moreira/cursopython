```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Definindo vari√°veis
x = [1,2,3]
y = [5,7,4]
x2 = [1,2,3]
y2 = [10,14,12]
```

```python
#Plotando linhas
plt.figure(figsize=(8,8)) #Definindo o tamanho da figura
plt.plot(x,y, label='Funcao 1') #Definindo a linha a ser plotada como label 'Funcao 1'
plt.plot(x2,y2, label='Funcao 2') #Definindo a linha a ser plotada como label 'Funcao 2'
plt.title('Titulo')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()
plt.savefig('plot.png') #Salvando figura
plt.show #O que isso faz?
```
